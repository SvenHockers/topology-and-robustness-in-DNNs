## Some examples of shape generation using the different methods

```python
# Unit circle in R^2 (closed curve)
circle = Shape.from_continuous(
    lambda t: (math.cos(2*math.pi*t), math.sin(2*math.pi*t)),
    filled=False 
)
circle.validate()
print(circle)
```

```python
square = Shape.from_polygon(
    [(0,0), (1,0), (1,1), (0,1)],
    filled=False  # boundary-only (no interior)
)
square.validate()
print(square)
```

```python
triangle = Shape.from_mesh(
    vertices=[(0,0), (1,0), (0,1)],
    cells=[(0,1,2)],
    filled=True
)
triangle.validate()
print(triangle)
```

```python
poly = Shape.create(
    dim=2,
    boundary=Boundary.POLYGON,
    filled=False,
    data={"vertices": [(0,0), (1,0), (1,1), (0,1)]}
)
poly.validate()
print(poly)
```
